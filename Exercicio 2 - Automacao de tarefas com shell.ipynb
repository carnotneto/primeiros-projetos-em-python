```python
#Exercício 2: Automação de Tarefas com Shell
#Crie um programa Python que execute um comando de shell (por exemplo, listagem de arquivos em um diretório) e capture a saída em uma variável.


import subprocess
comando = 'dir'
try:
    saida = subprocess.check_output(comando, shell=True, universal_newlines=True)
    print('Saída:')
    print(saida)
except subprocess.CalledProcessError as c:
    print(f'Erro ao executar o comando {c}')
except FileNotFoundError: 
    print('Comando não encontrado')
```
