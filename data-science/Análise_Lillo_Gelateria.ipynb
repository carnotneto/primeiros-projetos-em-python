```python
vendas = eventos_ofertas.groupby('cliente').sum()
vendas
```

```python
dados_clientes.rename(columns={'id': 'cliente'}, inplace=True)
dados_clientes
```

```python
merged = pd.merge(vendas, dados_clientes, on='cliente')
merged
```

```python
merged.isnull().sum()
```

```python
merged.nunique()
```

```python
columns_to_describe = ['tempo_decorrido','valor','idade', 'renda_anual']
merged[columns_to_describe].describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_subplots = len(columns_to_describe)
fig, axes = plt.subplots(1, num_subplots, figsize=(15, 6))
for i, column in enumerate(columns_to_describe):
    if column in columns_to_describe:
        merged[columns_to_describe].boxplot(column, ax=axes[i])
        axes[i].set_title(column)
plt.suptitle('Box Plots das colunas de interesse')
plt.tight_layout()
plt.show()
```
